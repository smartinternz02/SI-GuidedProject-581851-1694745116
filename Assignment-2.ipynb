```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("car_crashes")
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.scatterplot(x="speeding",y="alcohol",data=df)
```

Inference:we can say that as a speeding increases alcohol is slightly increasing

```python
sns.lineplot(x="speeding",y="not_distracted",data=df)
```

Inference: as a speeding increases there is a significant deviation in the people who are not_distracted

Inference - The "alcohol" variable appears skewed in nature and most of the bill values are in the range of 3 to 6

```python
sns.relplot(x="alcohol",y="speeding",data=df)  
```

```python

sns.barplot(data=df,x="alcohol",y="speeding")  
```

Inference: more alcohol value for high speeding

```python
sns.barplot(data=df,x="not_distracted",y="alcohol") 
```

Inference: more alcohol value for highly not_distracted

```python
sns.countplot(x="total",data=df)
```

```python
sns.jointplot(x="total",y="alcohol",data=df) 
```

scatterplot appears to be showing the strong correlation between the total & the alcohol. On the top of it, we can see the histogram of the respective variables.

## Box plot

```python
sns.boxplot(x="total",y="alcohol",data=df) 
```

Inference: there is significant deviation in the high total value of alchol
